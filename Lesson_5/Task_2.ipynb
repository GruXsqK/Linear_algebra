## Дополнительное практическое задание к уроку 5

```python
import numpy as np
from sklearn import datasets
from pandas import DataFrame
```

```python
iris = datasets.load_iris()
iris_frame = DataFrame(datasets.load_iris().data)
iris_frame.columns = iris.feature_names
iris_frame['target'] = iris.target
iris_frame['name'] = iris_frame.target.apply(lambda x : iris.target_names[x])
iris_frame.head(10)
```

Определим матрицу признаков

```python
A = np.array(iris_frame[['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)']])
print(f'Ранг матрицы признаков: {np.linalg.matrix_rank(A)}')
```

```python
U, s, V = np.linalg.svd(A)
print(f'Признаки:\n{s}')
print(f'Вес признаков:\n{[itm/sum(s) for itm in s]}')
```

Можем избавиться от 2 признаков, т.к. их вес в сумме менее 5%

```python
s[-2:] = 0
D = np.zeros_like(A, dtype=float)
D[np.diag_indices(min(A.shape))] = s
A_new = np.dot(np.dot(U, D), V)
print(f'Ранг новой матрицы признаков: {np.linalg.matrix_rank(A_new)}')
```

Найдем ковариационную матрицу, и сделаем сингулярное разложение

```python
A_cov = np.cov(A_new, rowvar=False)
U_cov, s_cov, V_cov = np.linalg.svd(A_cov)
print(f'Ковариационная матрица:\n{A_cov}')
```

```python
print(f'Получили матрицу перехода:\n{V_cov}')
```

Перейдем к новому базису с помощью матрицы перехода

```python
print(np.round(np.dot(V_cov, A_new.T).T, 10))
```
