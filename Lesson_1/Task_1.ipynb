## Практическое задание к уроку 1-1

__1.__ Исследовать на линейную зависимость:
$f_{1}(x)=e^{x}, f_{2}(x)=1, f_{3}(x)=x+1, f_{4}(x)=x-e^{x}.$

Заметим, что $f_{1}(x)=-f_{2}(x)+f_{3}(x)-f_{4}(x)$, то есть вектор $f_{1}(x)$ — линейная комбинация векторов $f_{2}(x)$, $f_{3}(x)$ и $f_{4}(x)$, из чего можно сделать вывод, что $f_{1}(x)=e^{x}$ и $f_{2}(x)=1$, $f_{3}(x)=x+1$, $f_{4}(x)=x-e^{x}$ линейно зависимы.

__2.__ Исследовать на линейную зависимость:
$f_{1}(x)=2, f_{2}(x)=x, f_{3}(x)=x^{2}, f_{4}(x)=(x+1)^{2}.$

Заметим, что $f_{4}(x)=f_{3}(x)+2\cdot f_{2}(x)+\frac{1}{2}\cdot f_{1}(x)$, то есть вектор $f_{4}(x)$ — линейная комбинация векторов $f_{1}(x)$, $f_{2}(x)$ и $f_{3}(x)$, из чего можно сделать вывод, что $f_{4}(x)=(x+1)^{2}$ и $f_{1}(x)=2$, $f_{2}(x)=x$, $f_{3}(x)=x^{2}$ линейно зависимы.

__3.__ Найти координаты вектора $x = (2, 3, 5)\in \mathbb{R}^{3}$ в базисе $b_{1}=(0, 0, 10)$, $b_{2}=(2, 0, 0)$, $b_{3}=(0, 1, 0)$.

$x=(2, 3, 5)=\frac{1}{2}\cdot (0, 0, 10) + (2, 0, 0) + 3\cdot(0, 1, 0)=\frac{1}{2}\cdot b_1 + b_2 + 3\cdot b_3$, т.е. координатами вектора $x$ в базисе $b_1, b_2, b_3$ являются $\frac{1}{2}$, $1$, $3$.

__4.__ Найти координаты вектора $3x^{2}-2x+2\in\mathbb{R}^{3}[x]$:

   а) в базисе $1$, $x$, $x^{2}$;

   б) в базисе $x^{2}$, $x-1$, $1$.

а) $3x^2-2x+2=3\cdot(0, 0, x^2) - 2\cdot(0, x, 0) + 2\cdot(1, 0, 0)$, т.е. координатами вектора $3x^2-2x+2$ в базисе $(1, x, x^2)$ являются $2$, $-2$, $3$.

б) $3x^2-2x+2=3\cdot(x^2, 0, 0) - 2\cdot(0, x-1, 0) + 0\cdot(0, 0, 1)$, т.е. координатами вектора $3x^2-2x+2$ в базисе $(x^2, x-1, 1)$ являются $3$, $-2$, $0$.
